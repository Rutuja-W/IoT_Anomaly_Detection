```python
sample_size = 200_000
data_sampled = data.sample(n=sample_size, random_state=42)
```

```python
X = data_sampled.drop('Label', axis=1)
y = data_sampled['Label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train/Test split (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define the hyperparameters for GridSearchCV
params = {
    'C': [0.1, 1, 10],  # Regularization strength
    'max_iter': [1000, 5000]  # Control iterations for convergence
}

# Initialize GridSearchCV with LinearSVC
start = time.time()

clf = GridSearchCV(LinearSVC(), params, cv=3, verbose=3, n_jobs=-1)  # n_jobs=-1 to use all CPU cores
clf.fit(X_train, Y_train)

end = time.time()

# Output results
print("Best parameters:", clf.best_params_)
print("Accuracy:", clf.score(X_test, Y_test))
print("Confusion Matrix:\n", confusion_matrix(Y_test, clf.predict(X_test)))
print("Classification Report:\n", classification_report(Y_test, clf.predict(X_test)))

print("Total time:", end - start, "seconds")
```
